```
def draw_error():
    log_data = np.load("./log_data.npy")
    fig = plt.figure()
    sub_plot = fig.add_subplot(1, 1, 1)
    xs = np.arange(log_data.size)
    ys = np.log(log_data)
    sub_plot.plot(xs, ys)
    plt.show()
    fig.clf()
draw_error()
```
