# Expand mnist data & save as png.
The variable `mnist_data` below should set as the place the MNIST data
lies in, where "MIST data" means the four files:
    `t10k-images-idx3-ubyte.gz`, `t10k-labels-idx1-ubyte.gz`,
    `train-images-idx3-ubyte.gz`, `train-labels-idx1-ubyte.gz`.
If you first use "MIST data", set `mnist_data` as you prefer. Then the
four file above will be downloaded automatically to the directory
you've chosen.  

```
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
mnist_data = "/home/ulily/tensorflow/tfmbt/MNIST_data/"
mnist = input_data.read_data_sets(mnist_data, one_hot=True)
for i in np.arange(20):
    mpimg.imsave(sv_loc + "outputs/mnist_" + str(i) + ".png",
                 mnist.test.images[i].reshape(28, 28),cmap="gray")
```

The command above should have generated 20 png files. Let's try to open
the 11th one of them.
